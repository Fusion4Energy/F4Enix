````python
# Import the mctal module
from f4enix.output.mctal import Mctal

# Parse the Mctal file
file = 'mctal'
mctal = Mctal(file)

# get a summary of the min and max errors across tallies
mctal.get_error_summary().sort_values(by='tally num')
````
